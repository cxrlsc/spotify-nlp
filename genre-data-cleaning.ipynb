```python
import pandas as pd 
import numpy as np

import re
import nltk

import spacy
from spacy_langdetect import LanguageDetector
```

```python
df = pd.read_csv('genre_lyrics.csv')
df.head(3)
```

```python
df.info()
```

```python
# Removing any duplicates 
df = df.drop_duplicates().reset_index()
df.info()
```

```python
df['artist'] = df['artist'].str.lower()
df['title'] = df['title'].str.lower()
df['lyrics'] = df['lyrics'].str.lower()
df['genre'] = df['genre'].str.lower()
df
```

```python
words = set(nltk.corpus.words.words())
```

```python
def clean_lyrics(test_str):
    
    test_str = str(test_str)
    # 0) Removing apostrophes for tokenization purposes
    test_str = test_str.lower().replace("'", "")
    
    # 1) Removing unecessary textual data 
    res = re.search(r'lyrics' , test_str)
    emb_res = re.search(r'\d*embed$', test_str)

    test_str = test_str[res.end():emb_res.start()]

    # 2) Removing any punctuation (except parantheses)
    test_str = re.sub(r'[.,\-?:!;]', '', test_str)
    
    test_str = test_str.replace("\n", " ")
    
    #test_str_out = " ".join(w for w in nltk.wordpunct_tokenize(test_str) if w.lower() in words)
    
    return test_str #_out

df['lyrics'] = df['lyrics'].apply(clean_lyrics)
```

```python
df
```

```python
from spacy.language import Language
from spacy_langdetect import LanguageDetector

def get_lang_detector(nlp, name):
    return LanguageDetector()

nlp = spacy.load("en_core_web_sm")
Language.factory("language_detector", func=get_lang_detector)
nlp.add_pipe('language_detector', last=True)

#nlp = spacy.load('en_core_web_sm')
#nlp.add_pipe(LanguageDetector(), name='language_detector', last=True)
#nlp.add_pipe('languagedet', name='language_detector', last=True)
```

```python
def remove_nonenglish(text):
    doc = nlp(text)
    # document level language detection. Think of it like average language of the document!
    english = (doc._.language['language']) == "en"
    # sentence level language detection
    
    if english:
        return text 
    else:
        return 'non-english'
```

```python
df['lyrics'] = df['lyrics'].apply(remove_nonenglish)
```

```python
df = df[df['lyrics'] != 'non-english']
df
```

```python
# Removing unecessary index row
df = df.reset_index().drop(columns=['index'])
df
```

```python
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop_words = stopwords.words("english")
```

```python
stop_words_lst = list(i.replace("'", "") for i in stop_words)
```

```python
def remove_stopwords(text):
    stop_words = set(stop_words_lst)
    word_tokens = word_tokenize(text)
    
    filtered_str = [w for w in word_tokens if not w.lower() in stop_words]
    
    return ' '.join(filtered_str)
```

```python
cleaned_df = english_cleaned_df
```

```python
cleaned_df['lyrics'] = english_cleaned_df['lyrics'].apply(remove_stopwords)
```
